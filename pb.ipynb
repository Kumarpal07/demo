##Concat

```python
df1=pd.DataFrame({
    'Name':['A','B','C','D'],
    'Value':[1,2,3,4]
})
df2=pd.DataFrame({
    'Name':['P','Q','D'],
    'Value':[5,6,7]
})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join="inner")
```

```python
pd.concat([df1,df2],join='inner',sort=True)
```

```python
empty=pd.DataFrame({})
print(empty)
```

```python
empty.shape
```

```python
pd.merge(df1,df2,how='outer',on='Name')
```
